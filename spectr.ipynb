```python
import numpy as np
import matplotlib.pyplot as plt


duration = 4  # длительность сигнала в секундах
amplitude = 0.4  # амплитуда  (в пределах: +-1.0)

fs = 60000
fs1 = 80000
n = 128
t = np.linspace(0, 1, n, endpoint=True)


timeSamples1 = np.arange(np.ceil(duration * fs1)) / fs1

signal = amplitude * np.sin(2 * np.pi * 261 * timeSamples1) 
signal2 = amplitude * np.sin(2 * np.pi * 293 * timeSamples1) 
signal3 = amplitude * np.sin(2 * np.pi * 329 * timeSamples1) 
signal4 = amplitude * np.sin(2 * np.pi * 349 * timeSamples1) 
signal5 = amplitude * np.sin(2 * np.pi * 392 * timeSamples1) 
signal6 = amplitude * np.sin(2 * np.pi * 440 * timeSamples1) 
signal7 = amplitude * np.sin(2 * np.pi * 493 * timeSamples1) 

signalSumm = signal + signal2 + signal3 + signal4 + signal5 + signal6 + signal7


plt.plot(signalSumm[:2000])
plt.title("sine tone")
plt.xlabel("time / milliseconds")
plt.show()

N = duration * fs1


spector = np.fft.rfft(signalSumm)
x_axis = 2 * np.fft.rfftfreq(N, 1/fs)/n

plt.plot(x_axis, np.abs(spector))
plt.show()

new_sig = np.fft.irfft(spector)
plt.plot(new_sig[:500])
plt.show()
```
